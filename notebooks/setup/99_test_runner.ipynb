# Test Runner — Databricks Lakehouse & Transformation

Runs all DEMO and WORKSHOP notebooks in sequence and reports PASS / FAIL per notebook.

| Cell | Scope |
|---|---|
| **Cell 2** | Setup check |
| **Cell 3** | Day 1 — DEMO notebooks |
| **Cell 4** | Day 1 — WORKSHOP notebooks |
| **Cell 5** | Day 2 — DEMO notebooks |
| **Cell 6** | Day 2 — WORKSHOP notebooks |
| **Cell 7** | Summary |

> **Prerequisites:** `00_pre_config.ipynb` and `00_setup.ipynb` must have been run at least once.
> Each notebook has a **timeout of 600 seconds** (10 min). Adjust `TIMEOUT` if needed.

```python
# =============================================================================
# CELL 2 — Setup and helper function
# =============================================================================
TIMEOUT = 600  # seconds per notebook

all_results = []

def run_notebooks(section_name, notebooks):
    """Run a list of notebooks and collect results."""
    results = []
    print("=" * 60)
    print(f"  {section_name}")
    print("=" * 60)
    
    for name, path in notebooks:
        try:
            dbutils.notebook.run(path, TIMEOUT)
            status = "PASS"
        except Exception as e:
            status = f"FAIL  →  {str(e)[:120]}"
        results.append((section_name, name, path, status))
        all_results.append((section_name, name, path, status))
        icon = "✅" if status == "PASS" else "❌"
        print(f"  {icon} {name:35s} {status}")
    
    passed = sum(1 for _, _, _, s in results if s == "PASS")
    print(f"\n  → {passed}/{len(results)} passed\n")
    return results
```

```python
# =============================================================================
# CELL 3 — Day 1 DEMOS
# =============================================================================
day1_demos = [
    ("01 Medallion Architecture",     "../Day1/Demo/01_medallion_architecture_demo"),
    ("02 Batch & Stream Ingestion",   "../Day1/Demo/02_batch_stream_ingestion_demo"),
    ("03 Lakeflow Pipelines",         "../Day1/Demo/03_lakeflow_pipelines_demo"),
    ("04 Orchestration",              "../Day1/Demo/04_orchestration_demo"),
]

run_notebooks("DAY 1 — DEMOS", day1_demos)
```

```python
# =============================================================================
# CELL 4 — Day 1 WORKSHOPS
# =============================================================================
day1_workshops = [
    ("01 Ingestion Workshop",         "../Day1/Workshop/01_batch_stream_ingestion_workshop"),
    ("02 Lakeflow Workshop",          "../Day1/Workshop/02_lakeflow_pipelines_workshop"),
]

run_notebooks("DAY 1 — WORKSHOPS", day1_workshops)
```

```python
# =============================================================================
# CELL 5 — Day 2 DEMOS
# =============================================================================
day2_demos = [
    ("01 Delta Advanced",             "../Day2/Demo/01_delta_advanced_demo"),
    ("02 Optimization",               "../Day2/Demo/02_optimization_demo"),
    ("03 Cost Management",            "../Day2/Demo/03_cost_management_demo"),
    ("04 Security & Governance",      "../Day2/Demo/04_security_governance_demo"),
]

run_notebooks("DAY 2 — DEMOS", day2_demos)
```

```python
# =============================================================================
# CELL 6 — Day 2 WORKSHOPS
# =============================================================================
day2_workshops = [
    ("01 Optimization Workshop",      "../Day2/Workshop/01_optimization_workshop"),
    ("02 Security Workshop",          "../Day2/Workshop/02_security_governance_workshop"),
]

run_notebooks("DAY 2 — WORKSHOPS", day2_workshops)
```

```python
# =============================================================================
# CELL 7 — SUMMARY
# =============================================================================
print("\n" + "=" * 60)
print("  FINAL SUMMARY")
print("=" * 60)

total_passed = sum(1 for _, _, _, s in all_results if s == "PASS")
total_failed = len(all_results) - total_passed

print(f"\n  Total: {total_passed}/{len(all_results)} passed  |  {total_failed} failed\n")

if total_failed:
    print("  ❌ FAILED NOTEBOOKS:")
    print("  " + "-" * 56)
    for section, name, path, status in all_results:
        if status != "PASS":
            print(f"  [{section}] {name}")
            print(f"    Path: {path}")
            print(f"    {status}")
            print()
else:
    print("  ✅ ALL NOTEBOOKS PASSED")

print("=" * 60)
```
